```python

file_path = "/content/drive/MyDrive/Case_Study_02/elektrik_veri.xlsx"

# Tüm sayfaları tek seferde oku
dfs = pd.read_excel(file_path, sheet_name=None, engine="openpyxl")

print("Sayfa isimleri:", list(dfs.keys()))

df_tahsilat = dfs["Tahsilat"]
df_tahsilat_1 = dfs["Tahsilat 1"]
df_tahakkuk = dfs["Tahakkuk"]
df_tahakkuk_1 = dfs["Tahakkuk 1"]
df_tahakkuk_2 = dfs["Tahakkuk 2"]

print(
    f"Tahsilat: {df_tahsilat.shape} | "
    f"Tahsilat 1: {df_tahsilat_1.shape} | "
    f"Tahakkuk: {df_tahakkuk.shape} | "
    f"Tahakkuk 1: {df_tahakkuk_1.shape} | "
    f"Tahakkuk 2: {df_tahakkuk_2.shape}"
)
```

```python
df_tahsilat.info()
df_tahakkuk.info()

df_tahakkuk.describe()
df_tahakkuk.head()
```

```python
#Benzersiz Müşteri Sayıları
print("Hamamözü:", df_tahakkuk['sozlesme_hesap_no'].nunique())
print("Gümüşhacıköy:", df_tahakkuk_1['sozlesme_hesap_no'].nunique())
print("Göynücek:", df_tahakkuk_2['sozlesme_hesap_no'].nunique())
```

```python
#Tahakkuk Verilerini Birleştirme
df_all = pd.concat([df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2])
print("Toplam kayıt:", df_all.shape)
```

```python
#Veri Kalitesi Kontrolü (eksik veri)
df_all.isna().sum()
```

```python
#Veri Kalitesi Kontrolü (negatif tüketim)
df_all[df_all['kwh'] < 0].head()
```

```python
#Veri Kalitesi Kontrolü (genel istatistik)
df_all['kwh'].describe()
```

```python
#Hesap Sınıfına Göre Tablo Oluşturma
hesap_stats = df_all.groupby("Hesap Sınıfı")["kwh"].agg(["mean","median","std","count"])
hesap_stats.sort_values("mean", ascending=False)
```
